# Load required libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Load the noisy image

```python
noisy_image = cv2.imread("noisy_image2.png")
```

# Apply filters

```python
# Gaussian blur
gaussian_blur = cv2.GaussianBlur(noisy_image, (7, 7), 0)

# Median filtering
median_filtered = cv2.medianBlur(noisy_image, 7)

# Non-local Means Denoising
fastNlMeans_filtered = cv2.fastNlMeansDenoisingColored(noisy_image, None, 10, 10, 7, 21)
```

# Display the results

```python
# Display the original, Gaussian blur, median filtered, and fastNlMeans denoised images
plt.figure(figsize=(15, 5))

plt.subplot(141)
plt.imshow(cv2.cvtColor(noisy_image, cv2.COLOR_BGR2RGB))
plt.title('Original')

plt.subplot(142) 
plt.imshow(cv2.cvtColor(median_filtered, cv2.COLOR_BGR2RGB)) 
plt.title('Median Filter')

plt.subplot(143) 
plt.imshow(cv2.cvtColor(gaussian_blur, cv2.COLOR_BGR2RGB)) 
plt.title('Gaussian Blur')

plt.subplot(144) 
plt.imshow(cv2.cvtColor(fastNlMeans_filtered, cv2.COLOR_BGR2RGB)) 
plt.title('fastNlMeans Denoised')

plt.tight_layout()
plt.show()
```
